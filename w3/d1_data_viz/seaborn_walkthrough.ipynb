```python
# import seaborn
import seaborn as sns

# import matplotlib
import matplotlib.pyplot as plt

# load the built-in tips dataset
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

## Regression Line With a Confidence Interval

```python
# a regression line with a confidence inetrval plot

sns.lmplot(x='total_bill', y='tip', data = tips)
plt.show()
```

Now if we want to have two regression lines in one chart – one for female tips and one for male tips – we can modify the code above and use the hue parameter.

```python
# add second line with 'hue'

sns.lmplot(x='total_bill',y='tip', data = tips, hue = 'sex')
plt.show()
```

On the other hand, we could also have one plot for each of the regression lines and write only one line of code

```python
sns.lmplot(x='total_bill',y='tip', data = tips, col = 'sex')
plt.show()
```

## Residual Plot
Another plot that is related to the regression line is a residual plot. We don't need to compute residuals manually and then plot the results

```python
sns.residplot(x='total_bill',y='tip', data = tips)
plt.show()
```

## Strip Plot

```python
sns.stripplot(x='day',y ='tip',data = tips, size =5, jitter = True)

# overwrite the original y-label from 'tip' to 'tip ($)'
plt.ylabel('tip ($)')
plt.show()
```

## Swarm Plot

```python
#distribution of attributes
sns.swarmplot(x='day',y='tip',data = tips, size = 3, hue = 'sex')

# overwrite the original y-label from 'tip' to 'tip ($)'
plt.ylabel('tip ($)')
plt.show()
```

## Boxplot + Violin Plot

```python
# create a grid with 1 row 2 columns
plt.subplot(1,2,1)

# the first plot
sns.boxplot(x='day',y='tip', data = tips)
plt.ylabel('tip ($)')

plt.subplot(1,2,2)

# the second plot
sns.violinplot(x='day',y='tip', data = tips)
plt.ylabel('tip ($)')
plt.tight_layout()
plt.show()
```

## Heatmap

```python
# compute correlations between features
df_corr = tips.corr()

# plot the correlations
sns.heatmap(df_corr)
plt.title('Correlation plot')
plt.show()
```
