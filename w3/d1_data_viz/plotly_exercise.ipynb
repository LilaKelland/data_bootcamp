### Task 1: Create a scatterplot with a 1000 random data points. 

- x-axis values should come from a normal distribution using np.random.randn(1000)

- y-axis values should come from a uniform distribution over [0,1) using np.random.rand(1000)

```python
# obtain the x and y values:

random_x = np.random.randn(1000)
random_y = np.random.rand(1000)
```

```python
# define a data variable
data = [go.Scatter(
            x = random_x,
            y = random_y,
            mode = 'markers',
        )]
```

```python
# define the layout, include a title and axis labels
# create the layout 
layout = go.Layout(
    title = 'Random Data Scatterplot',
        
    xaxis = dict(title = 'random_x'), # x-axis label
    yaxis = dict(title = 'random_y'), # y-axis label
#     hovermode ='closest' # handles multiple points landing on the same vertical
)

```

```python
# Create a fig from the data and layout. Then plot the fig
# create the figure
fig = go.Figure(data=data, layout=layout)
# fig.update_layout(
#     title={
#         'text': "Scatter Plot of Randomness ",
#         'y':0.25,
#         'x':0.5,
#         'xanchor': 'center',
#         'yanchor': 'top'})
fig.show()
```

```python
# import plotly.express as px
import plotly.graph_objects as go

df = pd.read_csv('https://raw.githubusercontent.com/Pierian-Data/Plotly-Dashboards-with-Dash/master/Data/abalone.csv') 

# data = go.Histogram(x=df.length, nbinsx=5)
# data = go.Histogram(x=df.length, xbins.size=0.02, xbins.start=0, xbins.end=1,)
data = go.Histogram(
            x=df.length,  
            xbins=dict(start=0, end=1, size=0.02)
        )

layout = go.Layout(
    title = "Abalone Length",
    xaxis = dict(title = 'length'), # x-axis label
    yaxis = dict(title = 'count') # y-axis label
)
fig = go.Figure(data=data, layout=layout)
# fig = go.Figure(data=go.Histogram(x=df.length, xbins=dict(start=0, end=1, size=0.02)), layout=layout)
fig.show()

```
