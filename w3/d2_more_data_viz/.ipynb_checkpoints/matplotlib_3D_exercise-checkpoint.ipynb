#### Task 1: Create wireframe of hyperbolic paraboloid on the domain <-10,10>X<-10,10>

* hyperbolic paraboloid equation is: z = x*x - y*y

```python

def function_z(x,y):
    return(x*x-y*y)
```

```python
N=50 # number of values
x_values = np.linspace(-10,10,N)
y_values = np.linspace(-10,10,N)

X,Y=np.meshgrid(x_values, y_values)
# X
```

```python
# calculate Z for everypoint
Z=function_z(X,Y)
Z.shape
```

```python
ax = plt.axes(projection='3d')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('hyperbolic paraboloid', fontsize=18)
ax.plot_wireframe(X,Y,Z);
```

#### Task 2: Create surface plot of eliptical paraboloid on the domain <-10,10>X<-10,10>

* eliptical paraboloid : z = x*x + y*y

```python
# surface plot

def function_z(x,y):
    return(x*x+y*y)
```

```python
N=50
x = np.linspace(-10,10,N)
y = np.linspace(-10,10,N)
X,Y= np.meshgrid(x,y)
```

```python
Z=function_z(X,Y)
```

```python
ax = plt.axes(projection='3d')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('eliptical  paraboloid', fontsize=18)
ax.plot_surface(X,Y,Z);
```

#### Task 3: Create surface plot of eliptical paraboloid on the domain <-10,10>X<-10,10>

* Also create a contour projection on the plane z=0 (HINT: use the parameter 'zdir' of contour plot)

* eliptical paraboloid : z = x*x + y*y

```python
# contour plot
def function_z(x,y):
    return(x*x + y*y)
```

```python

x = np.linspace(-10, 10, 80)
y = np.linspace(-10, 10, 80)

X, Y = np.meshgrid(x, y)
Z = function_z(X, Y)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour(X, Y, Z)
# ax.contour(X, Y, Z, zdir='z')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

ax.view_init(90, 0)
fig
```
