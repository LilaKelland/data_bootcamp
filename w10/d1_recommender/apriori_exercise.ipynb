```python
# import pandas
import pandas as pd  

# from apyori import apriori
from apyori import apriori
```

```python
# load data 
movie_data = pd.read_csv('https://raw.githubusercontent.com/pirandello/apriori/master/movie_dataset.csv', 
                         header = None)
```

* explore the dataset

```python
print(movie_data.shape)
movie_data.head(10)
```

* transform dataframe to list of lists (suitable format for apyori)

```python
records = []
for i in range(0, movie_data.shape[0]-1):
    records.append([str(movie_data.values[i,j]) for j in range(0, movie_data.shape[1])])
```

```python
records[0:2]
```

* instantiate apriori and set the input params based on following constraints:
    * we want only movies that are purchased at least 40 times
    * the minimum confidence for the rules is 20%
    * the minumum lift is 3

```python
association_rules = apriori(records, min_support=40/7500, min_confidence=0.2, min_lift=3, min_length=2)
association_results = list(association_rules)
```

* how many association rules did we obtained ?

```python
print(len(association_results))
```

* print the first association rule

```python
print(association_results[0:2])
```

* convert association rules to DataFrame 
    * use these columns: title_1, title_2, support, confidence, lift

```python
associations = pd.DataFrame(association_results)
associations
```

```python

associations['items'] = associations['items'].map(lambda x: list(x))

associations['title_1']= associations['items'].map(lambda x:x[0])
associations['title_2']= associations['items'].map(lambda x:x[1])
```

```python


associations['split_stats']=associations['ordered_statistics'].astype(str)
associations['split_stats']=associations['split_stats'].map(lambda x:x.split(sep = '='))

associations['confidence']= associations['split_stats'].map(lambda x:x[3]) #confidence
associations['lift'] =associations['split_stats'].map(lambda x:x[4]) #lift

# associations
```

```python
# OKAY NOW DROP REST AND STRIP REST WITH REGEX (lift - everything not a number of ., confidence eve after ,)
associations.drop(columns=['items', 'ordered_statistics','split_stats'], inplace = True)
associations
```
